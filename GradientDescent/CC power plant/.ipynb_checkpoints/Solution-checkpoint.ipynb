```python
import pandas as pd
boston_data=pd.read_csv("training_ccpp_x_y_train.csv")
boston_data
```

```python
x_train=boston_data.iloc[:,0:4]

x_train
```

```python
y_train=boston_data.iloc[:,4:5]
y_train
```

```python
import numpy as np
from sklearn.ensemble import GradientBoostingRegressor
model=GradientBoostingRegressor()
model.fit(x_train,y_train)
boston_data1=np.loadtxt('test_ccpp_x_test.csv',delimiter = ',')

y_predict=model.predict(boston_data1)

np.savetxt("y_boston2.csv", y_predict)
```
